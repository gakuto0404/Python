```python
# 例：１０１行目から１０５行目まで抽出
df[100:105]
```

```python
# Seriesをスライスでデータ抽出
df['人口（総数）'][:5]
```

```python
# 列名を指定したDataFrameをスライスでデータ抽出
df[['人口（総数）']][:5]
```

### 条件に一致する行のみ抽出

```python
# 条件に一致する行のみ抽出
df['西暦（年）']==2015
```

```python
# 表示
df[df['西暦（年）']==2015]
```

```python
# 文字列での表示
df[df['都道府県名'] == '東京都']
```

```python
df['西暦（年）'] % 10 == 0
```

```python
df[df['西暦（年）'] % 10 == 0]
```

```python
df[~(df['西暦（年）'] % 10 == 0)]
```

```python
df[df['西暦（年）'] % 10 != 0]
```

```python
df[(df['西暦（年）']==2015) & (df['都道府県名'] == '東京都')]
```

```python
df[(df['西暦（年）']==2010) | (df['西暦（年）']==2015)]
```
