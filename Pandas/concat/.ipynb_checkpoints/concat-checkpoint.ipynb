```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df01 = pd.DataFrame({'氏名':['高橋', '伊藤', '鈴木', '佐藤'],
                     'クラス':['df01', 'df01', 'df01', 'df01'],
                     '数学':[1, 2, 3, 4],
                     '国語':[5, 6, 7, 8]})
df01
```

```python
df02 = pd.DataFrame({'氏名':['高橋', '伊藤', '鈴木'],
                     'クラス':['df02', 'df02', 'df02'],
                     '数学':[9, 10, 11],
                     '国語':[12, 13, 14]})
df02
```

```python
pd.concat([df01, df02])
```

```python
pd.concat([df01, df02], ignore_index=True)
```

```python
df03 = pd.DataFrame({'氏名':['高橋', '伊藤', '坂本'],
                     'クラス':['df03', 'df03', 'df03'],
                     '数学':[15, 16, 17],
                     '国語':[18, 19, 20]})
df03
```

```python
pd.concat([df01, df02, df03])
```

```python
df = pd.concat([df01, df02, df03], keys=['一番目', '二番目', '三番目'])
```
