### 時系列データの集計

```python
# 売上日と売上金額のみのデータフレームを作成
df_03 = df[['売上金額']]
df_03
```

#### resampleメソッドは指定した頻度で集約する。（）の中に集計したい期間のコードを入れる

```python
df_03.resample('M')
```

```python
df_03.resample('M').sum() # 月ごとの合計
```

```python
df_03.resample('Q').sum() # 四半期ごとの合計
```

```python
df_03.resample('10D').sum() # 10日ごとの合計
```

```python
df_03.resample('M').mean() # 月ごとの平均
```

```python
df_03.resample('M').max() # 月ごとの最大
```

```python
df_03.resample('M').min() # 月ごとの最小
```

```python
df_03.resample('Q').agg(["sum", "max", "min", "mean"]) # 四半期ごとの合計・最大・最小・平均
```

#### ０が月曜日、６が日曜日

#### 曜日ごとの抽出と集計

```python
# 売上日の曜日を0〜6の数値で取得
df_03.index.weekday
```

```python
df_03[df_03.index.weekday == 0] # 条件抽出　月曜日のみ抽出
```

```python
df_03[df_03.index.weekday == 0].mean() # 月曜日のみの平均を算出
```

```python
# 各曜日ごとの合計を同時に算出
df_04 = df_03.set_index(df_03.index.weekday)
df_04.index.name = "曜日番号"
df_04.sum(level="曜日番号").sort_index()
```
